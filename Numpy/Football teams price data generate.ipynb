```python
import pandas as pd
import numpy as np
import random

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

# Number of samples
num_samples = 50000
```

```python
# Generate numeric columns with random data
numeric_data = {
    "AveragePlayerAge": np.random.uniform(18, 35, num_samples),
    "TotalGoalsLastSeason": np.random.randint(20, 100, num_samples),
    "MatchesWonLastSeason": np.random.randint(0, 38, num_samples),
    "MatchesDrawnLastSeason": np.random.randint(0, 38, num_samples),
    "MatchesLostLastSeason": np.random.randint(0, 38, num_samples),
    "TotalGoalsConcededLastSeason": np.random.randint(20, 100, num_samples),
    "TotalRevenueLastSeason": np.random.uniform(50, 500, num_samples),
    "StadiumCapacity": np.random.randint(10000, 100000, num_samples),
    "AverageAttendance": np.random.uniform(5000, 90000, num_samples),
    "TransferSpendingLastSeason": np.random.uniform(0, 200, num_samples),
    "TransferIncomeLastSeason": np.random.uniform(0, 200, num_samples),
    "NumberOfTrophies": np.random.randint(0, 50, num_samples),
    "MarketValueOfSquad": np.random.uniform(10, 1000, num_samples),
    "AveragePlayerMarketValue": np.random.uniform(0.5, 50, num_samples),
    "YouthAcademyRating": np.random.uniform(0, 100, num_samples),
}

# Generate categorical columns with random data
categorical_data = {
    "League": np.random.choice(["Premier League", "La Liga", "Bundesliga", "Serie A", "Ligue 1"], num_samples),
    "Country": np.random.choice(["England", "Spain", "Germany", "Italy", "France"], num_samples),
    "Manager": np.random.choice(["Manager A", "Manager B", "Manager C", "Manager D"], num_samples),
    "TeamFormation": np.random.choice(["4-4-2", "4-3-3", "3-5-2", "4-2-3-1", "3-4-3"], num_samples),
    "PlayingStyle": np.random.choice(["Attacking", "Defensive", "Possession", "Counter-Attack"], num_samples),
    "HomeCity": np.random.choice(["City A", "City B", "City C", "City D", "City E"], num_samples),
    "StadiumType": np.random.choice(["Modern", "Classic", "Multipurpose", "Historical"], num_samples),
    "MainSponsor": np.random.choice(["Sponsor A", "Sponsor B", "Sponsor C", "Sponsor D"], num_samples),
    "KitManufacturer": np.random.choice(["Manufacturer A", "Manufacturer B", "Manufacturer C"], num_samples),
    "OwnershipType": np.random.choice(["Private", "Public", "Consortium", "Fan-Owned"], num_samples),
}

# Generate target column (Price) with random data in the range 0.5 to 5 billion
price_data = {
    "Price": np.random.uniform(0.5, 5, num_samples) * 1e9
}

# Combine all data into a DataFrame
data = {**numeric_data, **categorical_data, **price_data}
df = pd.DataFrame(data)

# Format the Price column to show values in billions with commas and all zeros
df['Price'] = df['Price'].apply(lambda x: f"{x:,.0f}")

# Display the first few rows of the DataFrame
df.head()
```
