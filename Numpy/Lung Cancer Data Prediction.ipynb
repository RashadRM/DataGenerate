```python
import pandas as pd
import numpy as np
import random

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

# Number of samples
num_samples = 23658

# Generate data
patient_id = [f'Patient{i:04d}' for i in range(num_samples)]
age = np.random.randint(30, 80, size=num_samples)
gender = np.random.choice(['Male', 'Female'], size=num_samples)
smoking_history = np.random.choice(['Never Smoked', 'Former Smoker', 'Current Smoker'], size=num_samples)
tumor_size_mm = np.random.uniform(10, 100, size=num_samples)
tumor_location = np.random.choice(['Upper Lobe', 'Middle Lobe', 'Lower Lobe'], size=num_samples)
stage = np.random.choice(['Stage I', 'Stage II', 'Stage III', 'Stage IV'], size=num_samples)
treatment = np.random.choice(['Surgery', 'Chemotherapy', 'Radiation Therapy', 'Targeted Therapy'], size=num_samples)
survival_months = np.random.randint(1, 120, size=num_samples)

# Additional columns
ethnicity = np.random.choice(['Caucasian', 'African American', 'Asian', 'Hispanic', 'Other'], size=num_samples)
insurance_type = np.random.choice(['Private', 'Medicare', 'Medicaid', 'Other'], size=num_samples)
family_history = np.random.choice(['Yes', 'No'], size=num_samples)
comorbidity_diabetes = np.random.choice(['Yes', 'No'], size=num_samples)
comorbidity_hypertension = np.random.choice(['Yes', 'No'], size=num_samples)
comorbidity_heart_disease = np.random.choice(['Yes', 'No'], size=num_samples)
comorbidity_chronic_lung_disease = np.random.choice(['Yes', 'No'], size=num_samples)
comorbidity_kidney_disease = np.random.choice(['Yes', 'No'], size=num_samples)
comorbidity_autoimmune_disease = np.random.choice(['Yes', 'No'], size=num_samples)
comorbidity_other = np.random.choice(['Yes', 'No'], size=num_samples)
performance_status = np.random.choice(['0', '1', '2', '3', '4'], size=num_samples)
blood_pressure_systolic = np.random.randint(90, 180, size=num_samples)
blood_pressure_diastolic = np.random.randint(60, 110, size=num_samples)
blood_pressure_pulse = np.random.randint(60, 100, size=num_samples)
hemoglobin_level = np.random.uniform(10, 18, size=num_samples)
white_blood_cell_count = np.random.uniform(3.5, 10, size=num_samples)
platelet_count = np.random.uniform(150, 450, size=num_samples)
albumin_level = np.random.uniform(3, 5, size=num_samples)
alkaline_phosphatase_level = np.random.uniform(30, 120, size=num_samples)
alanine_aminotransferase_level = np.random.uniform(5, 40, size=num_samples)
aspartate_aminotransferase_level = np.random.uniform(10, 50, size=num_samples)
creatinine_level = np.random.uniform(0.5, 1.5, size=num_samples)
ldh_level = np.random.uniform(100, 250, size=num_samples)
calcium_level = np.random.uniform(8, 10.5, size=num_samples)
phosphorus_level = np.random.uniform(2.5, 5, size=num_samples)
glucose_level = np.random.uniform(70, 150, size=num_samples)
potassium_level = np.random.uniform(3.5, 5, size=num_samples)
sodium_level = np.random.uniform(135, 145, size=num_samples)
smoking_pack_years = np.random.uniform(0, 100, size=num_samples)

# Create DataFrame
data = {
    'Patient_ID': patient_id,
    'Age': age,
    'Gender': gender,
    'Smoking_History': smoking_history,
    'Tumor_Size_mm': tumor_size_mm,
    'Tumor_Location': tumor_location,
    'Stage': stage,
    'Treatment': treatment,
    'Survival_Months': survival_months,
    'Ethnicity': ethnicity,
    'Insurance_Type': insurance_type,
    'Family_History': family_history,
    'Comorbidity_Diabetes': comorbidity_diabetes,
    'Comorbidity_Hypertension': comorbidity_hypertension,
    'Comorbidity_Heart_Disease': comorbidity_heart_disease,
    'Comorbidity_Chronic_Lung_Disease': comorbidity_chronic_lung_disease,
    'Comorbidity_Kidney_Disease': comorbidity_kidney_disease,
    'Comorbidity_Autoimmune_Disease': comorbidity_autoimmune_disease,
    'Comorbidity_Other': comorbidity_other,
    'Performance_Status': performance_status,
    'Blood_Pressure_Systolic': blood_pressure_systolic,
    'Blood_Pressure_Diastolic': blood_pressure_diastolic,
    'Blood_Pressure_Pulse': blood_pressure_pulse,
    'Hemoglobin_Level': hemoglobin_level,
    'White_Blood_Cell_Count': white_blood_cell_count,
    'Platelet_Count': platelet_count,
    'Albumin_Level': albumin_level,
    'Alkaline_Phosphatase_Level': alkaline_phosphatase_level,
    'Alanine_Aminotransferase_Level': alanine_aminotransferase_level,
    'Aspartate_Aminotransferase_Level': aspartate_aminotransferase_level,
    'Creatinine_Level': creatinine_level,
    'LDH_Level': ldh_level,
    'Calcium_Level': calcium_level,
    'Phosphorus_Level': phosphorus_level,
    'Glucose_Level': glucose_level,
    'Potassium_Level': potassium_level,
    'Sodium_Level': sodium_level,
    'Smoking_Pack_Years': smoking_pack_years
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv(r'C:\Users\User\Desktop\Rashad\DATA\generated data\lung_cancer_data.csv', index=False)
```
